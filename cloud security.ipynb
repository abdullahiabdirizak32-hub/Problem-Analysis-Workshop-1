The following program displays a table with all the cloud security attack techniques, their detection methods and what solutions to be implemented

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Attack_Dataset.csv')

# Filter for Cloud security category
cloud_df = df[df['Category'].str.lower() == 'cloud security']

# Select relevant columns
columns = [
    'ID', 'Title', 'Category', 'Scenario Description',
    'Vulnerability', 'Detection Method', 'Solution'
]
cloud_table = cloud_df[columns]

# Display the table
cloud_table
```
